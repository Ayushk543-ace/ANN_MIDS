```python
import numpy as np
import matplotlib.pyplot as plt
```

***ANN Assignment 1

```python
class AF():
    def linear(x):
     return x

    def sigmoid(x):
        return 1/(1+np.exp(-x))
    def tanh(x):
        return np.tanh(x)
    def RELU(x):
        x1=[]
        for i in x:
            if i<0:
                x1.append(0)
            else:
                x1.append(i)
        return x1

    def softmax(x):
        return np.exp(x) / np.sum(np.exp(x),axis=0)
```

```python
while True:
    n = int (input("Enter 1.Linear\n 2.Sigmoid\n 3.Tanh\n 4.RELU\n 5.Softmax\n "))
    if n==0:
        break
    if n==1:
        x=np.linspace (-10,20)
        plt.plot(x,AF.linear(x))
        plt.title('Activation Function:Linear')
        plt.show()
    if n==2:
        x=np.linspace (-10,20)
        plt.plot(x,AF.sigmoid(x))
        plt.title('Activation Function:Sigmoid')
        plt.show()
    if n==3:
        x=np.linspace (-10,20)
        plt.plot(x,AF.tanh(x))
        plt.title('Activation Function:tanh')
        plt.show()
    if n==4:
        x=np.linspace (-10,20)
        plt.plot(x,AF.RELU(x))
        plt.title('Activation Function:RELU')
        plt.show()
    if n==5:
        x=np.linspace (-10,20)
        plt.plot(x,AF.softmax(x))
        plt.title('Activation Function:Softmax')
        plt.show()
```
